## Part-Of-Speech (POS) Tagging & The Human Tagger

### 1. What is POS Tagging?
POS Tagging is the process by which a single POS tag is assigned to each word, numeral, symbol, or punctuation in a text[cite: 2]. Each tag indicates the syntactic or grammatical role of that word in the sentence[cite: 6].

### 2. The Two Primary Classes
| Class Type | Description | Examples  |
| :--- | :--- | :--- |
| **Open (Lexical)** | Constant growth; new words added regularly. | Nouns (IBM, cat), Verbs (see), Adjectives (old), Adverbs (slowly). |
| **Closed (Functional)** | Fixed set of words; rarely changes. | Determiners (the), Conjunctions (and), Pronouns (he), Modals (can). |

### 3. Activity: The Human Tagger Challenge

#### **The Penn Treebank Standard**
The Penn Treebank tagset consists of **36 tags** and is considered the industry standard in NLP. 

* **Who uses it?** It is the backbone for major libraries like **NLTK** and **SpaCy**, and it was used to annotate the massive datasets (like the Wall Street Journal corpus) that helped train modern LLMs.
* **Simulator:** You can test various algorithms on this [Online POS Tagging Simulator](https://parts-of-speech.info/).

**How was the WSJ Corpus annotated?**
The Wall Street Journal corpus isn't just a collection of articles. Between 1989 and 1996, human annotators at the University of Pennsylvania manually verified the part-of-speech for every single word in approximately 2,500 news stories.

**The Process**: For each sentence, an initial automated tagger made a guess. A human expert then corrected those guesses using the 36-tag Penn Treebank set.

**The Result**: This "Gold Standard" dataset became the primary training material for the taggers used in software like NLTK today.

**Explore the Corpus**:
While the full 1-million-word corpus requires a paid license from the Linguistic Data Consortium (LDC), NLTK includes a 10% sample (about 100,000 words) for free educational use.

```python
import nltk
from nltk.corpus import treebank

# Download the free 10% sample of the Penn Treebank (WSJ)
nltk.download('treebank')

# List the file IDs in the sample (files starting with 'wsj_')
print(f"Sample files found: {treebank.fileids()[:5]}...")

# Access the human-annotated (Word, Tag) pairs for a specific file
file_id = 'wsj_0001.mrg'
tagged_words = treebank.tagged_words(file_id)

print(f"\n--- Human-Verified Tags for {file_id} ---")
for word, tag in tagged_words[:10]:
    print(f"{word:15} | {tag}")

    # EXERCISE: Find a word that is tagged as 'NNP' (Proper Noun) 
    # and compare it to its tag in a different sentence.
```

### 4. Activity: The Human vs. The Gold StandardYour Task: 

1. Manually tag the sentence below using the Penn Treebank tags (e.g., DT, JJ, NN, VBD).
2. Run the code cell to see how the NLTK automated tagger performs.
3. Then, we will "peek" into the Wall Street Journal (WSJ) corpus to see how humans tagged similar words in a real-world "Gold Standard" dataset.Sentence: "The smart student programmed the AI carefully."

```python
import nltk
from nltk import word_tokenize, pos_tag

# The 'punkt_tab' is the newer requirement for word_tokenize
nltk.download('punkt')
nltk.download('punkt_tab') 
nltk.download('averaged_perceptron_tagger_eng')

sentence_text = "The smart student programmed the AI carefully."

# Step 1: Tokenization
tokens = word_tokenize(sentence_text)

# Step 2: Automated Penn Treebank Tagging
automated_tags = pos_tag(tokens)

print("--- NLTK Automated Output ---")
for word, tag in automated_tags:
    print(f"{word:12} | {tag}")
```

**Examining the "Gold Standard" (WSJ Follow-up)**
Now, instead of just trusting the machine, query the human-annotated WSJ corpus to verify if the tags they saw above are consistent with the "Gold Standard" used by experts.

```python
from nltk.corpus import treebank
nltk.download('treebank')

# Access the WSJ 'Gold Standard' sample
wsj_sample = treebank.tagged_words()

# Define the words we want to verify from our sentence
target_words = ["The", "student", "programmed", "carefully"]

print(f"--- Verification: How humans tagged these in the WSJ Corpus ---")

for target in target_words:
    # Find examples of this word in the WSJ corpus
    examples = [(w, t) for w, t in wsj_sample if w.lower() == target.lower()]
    
    if examples:
        # Show the unique tags humans assigned to this word in the news
        unique_tags = set([t for w, t in examples])
        print(f"Word: '{target}' | Human Tags in WSJ: {unique_tags}")
    else:
        print(f"Word: '{target}' | Not found in the 10% sample.")

# DISCUSSION: 
# Does 'programmed' appear as VBD (past tense) in the news?
# Does 'student' ever have a tag other than NN (noun)?
```
