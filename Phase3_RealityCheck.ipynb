# Phase 3: The Reality Check
## Smoothing, Perplexity, and Evaluation

In Phase 2, we saw that $P(\text{Sam} | \text{I}) = 0.0$. If a test sentence contains a single sequence the model hasn't seen, the entire probability of that sentence becomes zero[cite: 115]. We can't even calculate Perplexity because we can't divide by zero!

### 1. The Solution: Add-1 (Laplace) Smoothing
We pretend we saw every possible word combination at least once more than we actually did. We add 1 to every count and adjust the denominator by the Vocabulary size $|V|$ to keep the probability mass the same.

**The Formula:**
$$
P_{Add-1}(w_i | w_{i-1}) = \frac{count(w_{i-1}, w_i) + 1}{count(w_{i-1}) + |V|}
$$

```python
from collections import Counter

corpus = "I am Sam . Sam I am . I do not like red eggs and ham ."
tokens = corpus.split()
vocab = set(tokens)
V = len(vocab)

unigram_counts = Counter(tokens)
bigrams = list(zip(tokens, tokens[1:]))
bigram_counts = Counter(bigrams)

def get_smoothed_prob(prev_word, current_word):
    # Applying the Add-1 formula [cite: 121]
    count_prev = unigram_counts[prev_word]
    count_bigram = bigram_counts[(prev_word, current_word)]
    
    smoothed_p = (count_bigram + 1) / (count_prev + V)
    return smoothed_p

print(f"Original P(Sam|I) = 0.0")
print(f"Smoothed P(Sam|I) = {get_smoothed_prob('I', 'Sam'):.4f}")
print(f"Notice it is no longer zero!")
```

### 2. Strategic Backoff and Interpolation
Sometimes, even smoothing isn't enough. If our Trigram evidence is weak, we 'Backoff' to a Bigram, and then to a Unigram. 

#### **The NLP Kitchen Metaphor**
Think of it like a chef: If the complex Trigram 'soup' is missing an ingredient, we use the Bigram 'broth', or just the Unigram 'spices'.

### 3. How do we know it's working? Perplexity
**Perplexity (PP)** is the standard intrinsic metric for Language Models. It measures how 'surprised' a model is by a test set. 
* **Lower Perplexity = Better Model**.
* Minimizing perplexity is mathematically equivalent to maximizing the probability of the test set.

**The Perplexity Intuition:** A model with a perplexity of 10 is as confused as if it had to choose randomly between 10 equally likely words at each step.

### 4. Final Lab: Extrinsic vs. Intrinsic
* **Intrinsic:** Perplexity (fast, math-based).
* **Extrinsic:** Performance on a real task like Speech Recognition or Machine Translation (slow, real-world).
